```julia
using LinearAlgebra
using Plots

# Oscillatory ODE: y' = i*y, y(0)=1
f(t,y) = im*y

# Step size and grid
h = 0.001
tspan = 0:h:2π
N = length(tspan)

# Initialize solution
y = zeros(ComplexF64, N)
y[1] = 1.0 + 0im

# Nordsieck vector coefficients: [c1, c2]
Y = zeros(ComplexF64, 2)

# Loop over steps
for n in 1:(N-1)
    tn = tspan[n]
    
    # Endpoint equation: y_{n+1} = y_n + 2*c2 => c2 = (y_{n+1} - y_n)/2
    # Implicit solve for y_{n+1}
    function residual(y_next)
        # c2 from endpoint condition
        c2 = (y_next - y[n])/2
        # midpoint theta=1/2
        y_mid = y[n] + Y[1]*sin(pi/2) + c2*(1 - cos(pi/2))
        dy_mid = (pi/2)*Y[1]/h + (pi/2)*c2/h  # derivative w.r.t t
        return dy_mid - f(tn + h/2, y_mid)
    end
    
    # Initial guess for y_{n+1}
    y_guess = y[n] + h * f(tn, y[n])
    
    # Solve implicitly using a simple fixed-point iteration
    y_next = y_guess
    for iter in 1:5
        res = residual(y_next)
        y_next -= res*0.5  # damping for stability
    end
    
    # Recover c2 from endpoint
    c2 = (y_next - y[n])/2
    # Solve for c1 using midpoint derivative
    y_mid = y[n] + c2*(1 - cos(pi/2))
    c1 = h * f(tn + h/2, y_mid) / (pi/2) - c2
    
    Y .= [c1, c2]
    y[n+1] = y_next
end

# Plot real and imaginary parts
plot(tspan, real.(y), label="Re(y)")
plot!(tspan, imag.(y), label="Im(y)")
xlabel!("t")
ylabel!("y(t)")
title!("Implicit BDF-style Nordsieck with sine-cosine basis")
```
